```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##pie chart-TOp 3 country that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.title("Top 3 Country that use Zomato")
```

Observation:- Zomato maximum records or transaction are from india
After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 4.4 ---> Very Good
3. When rating is between 3.5 to 3.9 ---> Good
4. When rating is between 3.0 to to 3.4 ---> Average
5. When rating is between 2.5 to 2.9 ---> Average
6. When rating is between 2.0 to 2.4 ---> Poor

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings,palette=['White','red','orange','yellow','green','green'])
```

# Observation :- 
    1. Not rated count is very high
    2. Maximum number of rating are between 2.5 to 3.4

```python
## couunt plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
